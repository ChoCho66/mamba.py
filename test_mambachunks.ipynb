```python
import torch

from mambapy.mamba import MambaConfig, MambaBlock
```

```python
B, L, D = 42, 100, 16

config = MambaConfig(d_model=D, n_layers=3)
model = MambaBlock(config)
```

```python
x = torch.randn(B, L, D)
```

```python
output_full = model(x)[0]
```

```python
#output_full_2, hiddens = model(x, [torch.zeros(B, 2*D, 16) for _ in range(config.n_layers)])
output_full_2, hiddens = model(x, torch.zeros(B, 2*D, 16))
```

```python
torch.allclose(output_full, output_full_2)
```

```python
#hiddens = [torch.zeros(B, 2*D, 16) for _ in range(config.n_layers)]
hiddens = torch.zeros(B, 2*D, 16)

chunk = x[:, 0:50]
output_c1, hiddens = model(chunk, hiddens)

chunk = x[:, 50:]
output_c2, _ = model(chunk, hiddens)

output_ct = torch.cat([output_c1, output_c2], dim=1)
```

```python
torch.allclose(output_full, output_ct, atol=0.1)
```

```python
torch.allclose(output_full[:, :50], output_c1)
```

```python
torch.allclose(output_full[:, 50:], output_c2)
```
